```python
yaml_content = """
train: /content/dataset/train/images
val: /content/dataset/valid/images
test: /content/dataset/test/images

names:
  0: look_away
  1: natural
  2: phone
  3: rub_eye
"""

with open("/content/data_filtered.yaml", "w") as f:
    f.write(yaml_content)
```

```python
import os

label_paths = [
    "/content/dataset/train/labels",
    "/content/dataset/valid/labels",
    "/content/dataset/test/labels"
]

for path in label_paths:
    for file in os.listdir(path):
        file_path = os.path.join(path, file)
        with open(file_path, "r") as f:
            lines = f.readlines()
        new_lines = [line for line in lines if int(line.split()[0]) < 4]  # keep only 0–3
        with open(file_path, "w") as f:
            f.writelines(new_lines)
```

```python
from ultralytics import YOLO

# Load pretrained YOLO model (use yolov8s.pt for better accuracy, yolov8n.pt for speed)
model = YOLO("yolov8n.pt")

# Train only on 0–3 classes
model.train(
    data="/content/data_filtered.yaml",
    epochs=50,
    imgsz=640,
    batch=16
)
```
